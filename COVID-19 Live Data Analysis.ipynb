```python
import requests
import pandas as pd

# Show all columns
pd.set_option('display.max_columns', None)

url = "https://disease.sh/v3/covid-19/countries"

response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
#store number  of total countries as array 
total_countries=np.arange(len(df['country']))
total_countries
```

```python
df['active']
```

```python
#print country with most number of cases
# idxmax >is index
print(*df.loc[df['cases'].idxmax()][['country', 'cases']])   
```

#### Total Cases Based On Country 

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 60), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_countries,df['cases'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_countries, df['country'])  
plt.xlim(1,max(df['cases'])+10) #Set X-axis Limit
plt.xlabel('Total Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

#### Active Number of Cases based on countries


```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 60), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_countries,df['active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8) )
    
plt.yticks(total_countries, df['country'])  
plt.xlim(1,max(df['active'])+100) #Set X-axis Limit
plt.xlabel('active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

#### Total Deaths of each Country

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 60), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_countries,df['deaths'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8) )
    
plt.yticks(total_countries, df['country'])  
plt.xlim(1,max(df['deaths'])+100) #Set X-axis Limit
plt.xlabel('Total deaths')  
plt.title('Corona Virus Cases')  
plt.show()  
```

#### Stack All the columns in the bar chart

```python
#set index to country name 
# drop the original country column and set as only index
df=df.set_index('country',drop=True)
```

```python
df.head()
```

```python
# remove a column named countryInfo
df = df.drop('countryInfo', axis=1)
```

```python
df.head()
```

```python
### Plotting Based on Stacking ,best function to plot 
# stacked=True>>> The bars are combined into one bar per row
df.plot.barh(stacked=True,figsize=(10,60))
```

#### TOP 10 countries by Death

```python
top10 = df.nlargest(10,'deaths')

plt.figure(figsize=(10,6))

plt.bar(top10['country'],
        top10['deaths'])

plt.xticks(rotation=45)

plt.title("Top 10 Countries by Deaths")

plt.show()
```
